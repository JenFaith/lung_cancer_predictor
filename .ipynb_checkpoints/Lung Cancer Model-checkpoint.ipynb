## Part 1: Prepare Images by Removing Medical Text

```python
# Pathways to images
PATH = r'Images/'
PATH_CANCEROUS = r'Images/Cancerous/'
PATH_NON = r'Images/Non-Cancerous/'
```

As you can see from the embedded image, one of the issues with the non-cancerous images is that they all contain text. That text needs to be removed so as not to throw off the predictions.

```python
from IPython import display 
display.Image('Images/Non-Cancerous/Non-Cancerous14.jpg')
```

```python
def crop(directory):
    for file in os.listdir(directory):
        f_img = directory+"/"+file
        im = Image.open(f_img)

        # Cropping Image
        img = im.crop((0, 100, 512, 400))
        
        # Save Image
        img.save(f_img)
    return
```

```python
im = Image.open('Images/Non-Cancerous/Non-Cancerous14.jpg') 
width, height = im.size
width, height
```

```python
# Apply same crop to both cancerous and noncancerous images
crop(PATH_CANCEROUS)
crop(PATH_NON)
```

```python
# Define Image generators for training
train_datagen2 = ImageDataGenerator(
    featurewise_center=True, 
    featurewise_std_normalization=True,
    horizontal_flip=True,
    shear_range=0.2,
    validation_split = 0.25,
    zoom_range=0.2)
```

```python
# Define Image generators for testing
test_datagen2 = ImageDataGenerator(
    featurewise_center=True, 
    featurewise_std_normalization=True,
    horizontal_flip=True,
    shear_range=0.2,
    validation_split = 0.25,
    zoom_range=0.2)
```

```python
batch_size = 16
epochs = 30
```

```python
train = train_datagen2.flow_from_directory(batch_size=batch_size,
                                                     directory=PATH,
                                                     shuffle=True,
                                                     target_size=(300, 512),
                                                     class_mode='binary',
                                                     seed = 48,
                                                     subset="training",
                                                    )
```

```python
test = test_datagen2.flow_from_directory(batch_size=batch_size,
                                                     directory=PATH,
                                                     shuffle=True,
                                                     target_size=(300, 512),
                                                     class_mode='binary',
                                                     seed = 48,
                                                     subset="validation",
                                                    )
```

## Make the Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Conv2D, MaxPooling2D, Flatten
from keras.preprocessing.image import array_to_img, img_to_array, load_img
```

```python
# Setup

# random value that will get updated during Gradient Descent just like weights in the FCFF network 
n_weight_matrices = 32

# specify the window size (i.e. 3 cell high and 3 cell wide)
weight_matrix_size = (3,3)

act_func = 'relu'

# dim of the image: 32 cell high, 32 cell wide, and 3 channels (one for Red, one for Blue, and one for Green)
# this means our images are not matrices, they are tensors 
image_dim = (32, 32, 3)

pool_size = (2, 2)

model = Sequential()

# 1st conv layer 
model.add(Conv2D(n_weight_matrices, 
                 weight_matrix_size, 
                 activation=act_func, 
                 input_shape=image_dim))

# 1st pooling layer 
model.add(MaxPooling2D(pool_size))

# 2st conv layer 
model.add(Conv2D(64, weight_matrix_size, activation=act_func))

# 2st pooling layer 
model.add(MaxPooling2D(pool_size))

# 3st conv layer 
model.add(Conv2D(62, weight_matrix_size, activation=act_func))

# flatten the image matrix into a row vector 
model.add(Flatten())

# hidden layer in FCFF portion of model 
model.add(Dense(64, activation=act_func))

# Output layer 
model.add(Dense(units=1, activation="softmax"))

model.summary()
```

```python
# Compile Model
model.compile(optimizer='adam',
              loss='BinaryCrossentropy',
              metrics=['categorical_accuracy'])
```
